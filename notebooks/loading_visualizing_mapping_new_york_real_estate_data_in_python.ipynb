# Download New York City real estate data

PLUTO is a dataset containing information on every property in New York City. The data is collected by the Department of City Planning and is available in a publicly accessible database. The data is available in a variety of formats, including shapefiles, CSVs, and Excel files. The data is available for download from the NYC Open Data portal.

```python
!wget https://s-media.nyc.gov/agencies/dcp/assets/files/zip/data-tools/bytes/nyc_mappluto_24v1_fgdb.zip -O /tmp/nyc_mappluto_24v1_fgdb.zip
!unzip -o /tmp/nyc_mappluto_24v1_fgdb.zip -d ~/data/nyc_mappluto_24v1_fgdb
```

## Loading New York City real estate and property data in Python using GeoPandas

This takes ~15 minutes and maybe longer...

```python
import geopandas as gpd

file_path = "/Users/me/data/nyc_mappluto_24v1_fgdb/MapPLUTO24v1.gdb"
gdf = gpd.read_file(file_path)
```

```python
gdf
```

```python
gdf.plot(figsize=(10,10))
```

```python
# Convert the coordinate system to WGS84/EPSG:4326
gdf = gdf.to_crs("EPSG:4326")

# Convert to Parquet
gdf.to_parquet("/Users/me/data/nyc_mappluto_24v1_fgdb/MapPLUTO24v1_wgs84.parquet")
```

# Convert New York City real estate data to GeoJSON

```python
!brew install planetlabs/tap/gpq
```
