## Transform the coordinates

```python
!duckdb -c "INSTALL spatial; LOAD spatial; DESCRIBE SELECT * FROM ST_READ('/Users/me/data/nyc_mappluto_24v1_fgdb/MapPLUTO24v1.gdb') LIMIT 10;"
```

```python
%%sql 
SELECT 
    ST_Point(Latitude, Longitude) AS point,
    ST_Transform(point, 'EPSG:2263', 'EPSG:4326') AS point_4326
FROM ST_READ('/Users/me/data/nyc_mappluto_24v1_fgdb/MapPLUTO24v1.gdb') LIMIT 10;
```
